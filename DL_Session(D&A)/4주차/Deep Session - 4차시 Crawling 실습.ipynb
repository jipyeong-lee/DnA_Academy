## 1. Requests & BeautifulSoup4

requests의 get으로 url을 불러오면 text로 받아온 페이지의 html코드를 볼 수 있지만  text는 문자열객체를 반환할 뿐이기 때문에 requests는 불러올때 쓰고

이후 html 코드를 Python이 이해하는 객체 구조로 변환하고 정보를 추출해 내기 위해 BeautifulSoup을 사용한다

```python
import requests
from bs4 import BeautifulSoup
```

<font color=blue> **Yes24에서 책 정보 불러오기**

```python
# yes24 베스트셀러 페이지 url을 변수 raw에 저장
req = requests.get('http://www.yes24.com/24/Category/BestSeller')
print(req)
    # Response 200 출력시 정상으로 url을 불러왔다는 것
    # Response 400번대 출력시 어떤 오류로 정상적으로 url을 불러오지 못한 것
```

```python
# req.text로 해당 페이지의 모든 html 정보가 담아져 있음을 볼 수 있음
html = req.text
print(html)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
print(soup)
```

```python
# 불러온 html이 단순 텍스트가 아닌 태그로 감싸진 html 구조를 인지하여 html관점으로 불러옴
# 첫 인자는 html 소스코드, 두번째 인자는 어떤 parser를 이용할지 명시
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

똑같은 태그가 여러 개가 존재하는 것들은 대부분 class나 id로 구분함
 - id: `#`을 사용해 정보 찾기 ex) 태그#id
 - class: `.`을 사용해 정보 찾기 ex) 태그.class

태그는 상하관계가 있기 때무넹 찾고자 하는 여러 태그가 존재한다면 이를 이용
 - 자식 관계: 바로 한 단계 아래 하위태그. ` > ` 사용
 - 자손 관계: 모든 하위 태그. `띄어쓰기` 사용

select메소드 & select_one메소드
 - select: 주어진 태그에 해당하는 태그 정보를 `모두` 불러옴
 - select_one: 주어진 태그에 해당하는 정보를 `하나`만 불러옴

```python
books = soup.select('div#bestList ol > li')
print(books)
```

```python
# ol의 모든 하위태그 li -> 자식과 자손 태그 모두 불러옴
len(soup.select('div#bestList ol li'))
```

```python
# ol의 바로 한 단계 아래 하위태그 li -> 자식 태그 불러옴
len(soup.select('div#bestList ol > li'))
```

```python
book = soup.select_one('div#bestList ol > li')
print(book)
```

**검색하기**

```python
keyword = 'python'
```

```python
req = requests.get('http://www.yes24.com/searchcorner/Search?keywordAd=&keyword=&domain=ALL&qdomain=%C0%FC%C3%BC&query=' + keyword)
print(req)
```

```python
keywords = ['python', 'deep learning', 'sql']
```

```python
for keyword in keywords:
    print(f'------------------ {keyword} ------------------')
    req = requests.get('http://www.yes24.com/searchcorner/Search?keywordAd=&keyword=&domain=ALL&qdomain=%C0%FC%C3%BC&query=' + keyword)
    soup = BeautifulSoup(req.text, 'html.parser')
    books = soup.select('td.goods_infogrp')
    for book in books[:20]:
        title = book.select_one('p.goods_name a').text
        writer = book.select_one('div.goods_info a').text
        price = book.select_one('div.goods_price em').text
        print(f'{title}, {writer}, {price}')
```

```python
# 검색하고 페이지를 넘어가면서 끌어온다
for keyword in keywords:
    print(f'\n------------------ {keyword} ------------------')
    for page_num in range(1, 2):
        req = requests.get(
            f'http://www.yes24.com/searchcorner/Search?keywordAd=&keyword=&domain=ALL&qdomain=%c0%fc%c3%bc&query={keyword}&PageNumber={page_num}&scode=012')
        soup = BeautifulSoup(req.text, 'html.parser')
        books = soup.select('td.goods_infogrp')[:20]
        for book in books:
            title = book.select_one('a').text
            writer = book.select_one('div.goods_info a').text
            price = book.select_one('div.goods_price em').text
            print(f'{title}, {writer}, {price}')
            
```

<font color=blue> **네이버영화에서 영화 정보 불러오기**

```python
req = requests.get('https://movie.naver.com/movie/running/current.nhn', headers={'User_Agent':'Mozilla/5.0'})

soup = BeautifulSoup(req.text, 'html.parser')

soup
```

영화의 제목 / 장르 /배우를 스크랩 해본다

**같은 이름의 태그일 경우 `nth-of-type(num)`로 구별할 수 있다**  
 class나 id 명이 같을 때 뿐만 아니라 태그명이 겹칠 때도 사용할 수 있다

```python
movies = soup.select('dl.lst_dsc')

for movie in movies:
    title = movie.select_one('dt.tit a').text
    genre = movie.select_one('dl.info_txt1 dd a').text
    actor = movie.select_one('dl.info_txt1 dd a').text
    print(f'{title}, {genre}, {actor}')
    
```

```python
movies = soup.select('dl.lst_dsc')

for movie in movies:
    title = movie.select_one('dt.tit a').text
    genre = movie.select_one('dl.info_txt1 dd:nth-of-type(1) a').text
    actor = movie.select_one('dl.info_txt1 dd:nth-of-type(3) a').text
    print(f'{title}, {genre}, {actor}')
    
```

```python
movies = soup.select('dl.lst_dsc')

for movie in movies:
    title = movie.select_one('dt.tit a').text
    try:
        genre = movie.select_one('dl.info_txt1 dd:nth-of-type(1) a').text
    except:
        genre = ''
    try:
        actor = movie.select_one('dl.info_txt1 dd:nth-of-type(3) a').text
    except:
        actor = ''
    print(f'{title}, {genre}, {actor}')
    
```

```python
# 검색입력
keyword = dr.find_element_by_css_selector('div.green_window input')
keyword.send_keys('파이토치')
```

```python
# 클릭하기
search = dr.find_element_by_css_selector('button#search_btn')
search.click()
```

```python
books = dr.find_elements_by_css_selector('li.book_group')

len(books)
```

```python
for book in books:
    title = book.find_element_by_css_selector('div.book_info a').text
    writer = book.find_element_by_css_selector('div.item_info.type_writer').text
    view = book.find_element_by_css_selector('div.item:nth-of-type(3) a').text
    print(f'{title}, {writer}, {view}')
```

```python
for i in range(len(books)):
    title = books[i].find_element_by_css_selector('div.book_info a').text
    writer = books[i].find_element_by_css_selector('div.item_info.type_writer').text
    view = books[i].find_element_by_css_selector('div.item:nth-of-type(3) a').text
    print(f'{title}, {writer}, {view}')
    if (i+1) % 3 == 0:
        next_arrow = dr.find_element_by_css_selector('div.cmm_pgs a:nth-of-type(2)')
        next_arrow.click()        
```
