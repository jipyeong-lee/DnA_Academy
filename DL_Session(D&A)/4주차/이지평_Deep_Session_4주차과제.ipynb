```python
keywords = ['python', 'deep learning', 'machine learning', 'computer', 'math']
# 검색하고 페이지를 넘어가면서 끌어온다
for keyword in keywords:
    print(f'\n------------------ {keyword} ------------------')
    for page_num in range(1, 3):
        req = requests.get(
            f'http://www.yes24.com/searchcorner/Search?keywordAd=&keyword=&domain=ALL&qdomain=%c0%fc%c3%bc&query={keyword}&PageNumber={page_num}&scode=012')
        soup = BeautifulSoup(req.text, 'html.parser')
        books = soup.select('td.goods_infogrp')[:20]
        for book in books:
            title = book.select_one('p.goods_name a').text
            try:
                writers = book.select_one('div.goods_info').text.strip()
                writer = writers.split('|')[0].strip()
                company = writers.split('|')[1].strip()
                publication = writers.split('|')[2].strip()
            except:
                writers = '정보없음' 
                writer = '정보없음' 
                company = '정보없음' 
                publication = '정보없음' 
            try:
                price = book.select_one('div.goods_price em').text
            except:
                price = '정보없음'
            try:
                sell_count = book.select_one('div.goods_rating span').text.split('|')[0].strip()
                sell_count = str(sell_count)
                sell_count = re.findall('[0-9]', sell_count)
                sell_count = ('').join(sell_count)
            except:
                sell_count = '정보없음'
            try:
                review_count = book.select_one('div.goods_rating span a').text
                review_count = str(review_count)
                review_count = re.findall('[0-9]', review_count)
                review_count = ('').join(review_count)
            except:
                review_count = '정보없음'
            try:
                score = book.select_one('span.gd_rating em').text
            except:
                score = '정보없음'
            site_link = book.select_one('p.goods_name a').attrs['href']
            site_link = 'http://www.yes24.com' + book.select_one('p.goods_name a').attrs['href']
            
            print(f'{title}, {writer}, {company}, {publication}, {price}, {sell_count}, {review_count}, {score}, {site_link}')
```

```python
import time
path = 'C:/Users/leeji/ML_//chromedriver_win32/'

keywords = ['python', 'deep learning', 'machine learning', 'computer', 'math']

# 검색하고 페이지를 넘어가면서 끌어온다
for keyword in keywords:
    print(f'\n------------------ {keyword} ------------------')
    dr = webdriver.Chrome(path + 'chromedriver.exe')
    dr.get('http://www.yes24.com/main/default.aspx')
    kwd = dr.find_element_by_css_selector('span.schIpt label input')
    kwd.send_keys(keyword)    
    search = dr.find_element_by_css_selector('span.schBtn button')
    search.click()
    books = dr.find_elements_by_css_selector('td.goods_infogrp')[:20]
        
    for p in range(1, 3):
        for i in range(len(books)):
            title = books[i].find_element_by_css_selector('p.goods_name a').text
            try:
                writer = books[i].find_element_by_css_selector('div.goods_info a').text
            except:
                writer = '정보없음'
            try:
                price = books[i].find_element_by_css_selector('div.goods_price em').text
            except:
                price = '정보없음'
            try:
                sell_count = books[i].find_element_by_css_selector('div.goods_rating span').text
                sell_count = str(sell_count)
            except:
                sell_count = '정보없음'
            try:
                review_count = books[i].find_element_by_css_selector('div.goods_rating span a').text
                #review_count = 
            except:
                review_count = '정보없음'
            try:
                score = books[i].find_element_by_css_selector('span.gd_rating em').text
            except:
                score = '정보없음'
            site_link = books[i].find_element_by_css_selector('p.goods_name a').get_attribute('href')

            print(f'{title}, {writer}, {price}, {sell_count}, {review_count}, {score}, {site_link}')

            if (i == 20):
                next_arrow = dr.find_element_by_css_selector('div.pagen a.n')
                next_arrow.click()
                time.sleep(1)
                p += 1
            
    dr.close()
```

```python
def book_crawling_200():
    
    # 책 키워드 입력
    keywords = ['python', 'deep learning', 'machine learning', 'computer', 'math']
    finish = False
    
    # csv 폴더
    if not os.path.isdir('book_csv파일'):
        os.mkdir('book_csv파일')
    
    # xlsx 폴더
    if not os.path.isdir('book_xlsx파일'):
        os.mkdir('book_xlsx파일')

    # 폴더 없을 경우 폴더 생성
    for keyword in keywords:
            
        # 파일 생성_csv파일
        f = open(f'book_csv파일\{keyword}_csv파일.csv', 'w') 
        f.write('title,writer,company,publication,price,sell_count,review_count,score,site_link') # csv파일이라 쉼표로 셀 구분

        # 파일 생성_xlsx파일
        wb = openpyxl.Workbook()
        sheet = wb.active
        sheet.append(['title', 'writer', 'company', 'publication', 'price', 'sell_count', 'review_count', 'score', 'site_link'])
        
        # 사이트 열기
        dr = webdriver.Chrome(path + 'chromedriver.exe')
        dr.get('https://www.yes24.com')

        # 검색 키워드 입력
        kwd = dr.find_element_by_css_selector('.schIpt label input')
        kwd.send_keys(keyword)

        # 검색
        search = dr.find_element_by_css_selector('span.schBtn > button')
        search.click()

        # 무시하고 보내기ㅣ 클릭
        click = dr.find_element_by_css_selector('button#proceed-button')
        click.click()

        book_num = 1
        
        limit = int(dr.find_elements_by_css_selector('h2 em')[3].text.replace(',',''))
        
        for i in range(1, 3):
            books = dr.find_elements_by_css_selector('td.goods_infogrp')
            for x in range(0, 20):
            # 책 제목
                try:
                    title = books[x].find_element_by_css_selector('p.goods_name a').text
                except:
                    title = '없음'         
            # 저자    
                try:
                    writers = books[x].find_element_by_css_selector('div.goods_info').text.strip()
                    writer = writers.split('|')[0].strip()
                    company = writers.split('|')[1].strip()
                    publication = writers.split('|')[2].strip()
                except:
                    writers = '없음'
                    writer = '없음'
                    company = '없음'
                    publication = '없음'
            # 책 가격
                try:
                    price = int(books[x].find_element_by_css_selector('div.goods_price em').text)
                except:
                    price = 0
            # 판매 지수   
                try:
                    sell_count = books[x].find_element_by_css_selector('div.goods_rating span').text.split('|')[0].strip()
                    sell_count = str(sell_count)
                    sell_count = re.findall('[0-9]', sell_count)
                    sell_count = ('').join(sell_count)
                except:
                    sell_count = 0

            # 리뷰 수
                try:
                    review_count = books[x].find_element_by_css_selector('div.goods_rating span a').text
                    review_count = str(review_count)
                    review_count = re.findall('[0-9]', review_count)
                    review_count = int(('').join(review_count))
                except:
                    review_count = 0

            # 평점   
                try:
                    score = books[x].find_element_by_css_selector('span.gd_rating em').text
                except:
                    score = 0

            # 사이트 링크    
                site_link = books[x].find_element_by_css_selector('p.goods_name a').get_attribute('href')
                site_link = 'http://www.yes24.com' + site_link
                    
            # 저장
                sheet.append([title, writer, company, publication, price, sell_count, review_count, score, site_link])
                f.write('\n' + str(title).replace(',','') +','
                        +str(writer).replace(',','')+','
                        +str(company).replace(',','')+','
                        +str(publication).replace(',','')+','
                        +str(price).replace(',','')+','
                        +str(sell_count).replace(',','')+','
                        +str(review_count).replace(',','')+','
                        +str(score).replace(',','')+','
                        +str(site_link).replace(',',''))                
                book_num += 1
                
            # 검색 결과 초과시 종료
                if limit < book_num:
                    finish = True
                    break
                    
            if finish == True:
                break

            # 페이지 넘기기 
            pages = dr.find_element_by_css_selector(f'.pagen.pat30.pab15 a:nth-of-type({i})')
            pages.click()
        
        f.close()
        dr.close()
        
        #xlsx 저장
        wb.save(f'book_xlsx파일\{keyword}_xlsx파일.xlsx')
    print('----------------데이터 수집 완료--------------------')
```
