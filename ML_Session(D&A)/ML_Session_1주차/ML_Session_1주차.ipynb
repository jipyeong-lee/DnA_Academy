```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('titanic.csv') ; df
```

## 데이터 전처리

```python
df.loc[df['Sex'] == 'male','Sex'] = 0
df.loc[df['Sex'] == 'female','Sex'] = 1
```

```python
df.Age = df.Age.fillna(df['Age'].mean())
```

```python
df.Embarked = df.Embarked.fillna('S')
```

```python
df.loc[df['Embarked'] == 'S','Embarked'] = 0
df.loc[df['Embarked'] == 'C','Embarked'] = 1
df.loc[df['Embarked'] == 'Q','Embarked'] = 2
```

```python
df = df.drop(columns=['PassengerId','Name','Ticket','Cabin'])
df
```

## 데이터 분리

```python
X = df.drop(columns=['Survived'])
y = df.Survived
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
print('X_train :',X_train.shape)
print('y_train :',y_train.shape)
print('X_test :',X_test.shape)
print('y_test :',y_test.shape)
```

## 모델 학습

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=0)
```

```python
model.fit(X_train,y_train)
```

```python
print('예측값 :\n',model.predict(X_test))
```

```python
print('예측값 : \n',model.predict_proba(X_test))
```

## 모델 평가

```python
print('모델 학습 점수 :',model.score(X_train,y_train))
```

```python
print('테스트 점수 :',model.score(X_test,y_test))
```

## 다시 데이터 전처리

```python
df = pd.read_csv('titanic.csv')
```

```python
df.loc[df['Sex'] == 'male','Sex'] = 0
df.loc[df['Sex'] == 'female','Sex'] = 1
```

```python
df.Age = df.Age.fillna(df.Age.mean())
```

```python
df.loc[df['Age'] < 10,'Age'] = 0
df.loc[(df['Age']) >= 10 & (df['Age'] < 20),'Age'] = 1
df.loc[(df['Age']) >= 20 & (df['Age'] < 30),'Age'] = 2
df.loc[(df['Age']) >= 30 & (df['Age'] < 40),'Age'] = 3
df.loc[(df['Age']) >= 40 & (df['Age'] < 50),'Age'] = 4
df.loc[df['Age'] >= 50,'Age'] = 5
```

```python
df['Fare'] = df['Fare'].fillna(df.groupby('Pclass')['Fare'].transform('mean'))
df.head()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
df.head()
```

```python
df.Embarked = df.Embarked.fillna('S')
```

```python
df.loc[df['Embarked']=='S','Embarked'] = 0
df.loc[df['Embarked']=='C','Embarked'] = 1
df.loc[df['Embarked']=='Q','Embarked'] = 2
```

```python
df['rich_female'] = 0

df.loc[(df['Pclass']<=2) & (df['Sex'] == 'feamle'),'rich_female'] = 1
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Fare'])
df
```

## 다시 데이터 분리

```python
X = df.drop(columns=['Survived'])
y = df.Survived

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

## 모델 학습

```python
model = DecisionTreeClassifier(random_state=0)
```

```python
model.fit(X_train,y_train)
```

## 모델 평가

```python
print('모델 학습 점수 :',model.score(X_train,y_train))
```

```python
print('테스트 점수 :',model.score(X_test,y_test))
```
